```python
import numpy as np
import pandas as pd
import datetime

def suspect_no():
    return np.random.randint(1000000000,9999999999,dtype=np.int64)

def phone_no():
    freq=np.random.normal(3,4,52)
    freq=freq.astype(int)
    freq=freq[freq>0]
    phoneno=np.random.randint(6000000000,9999999999,25,dtype=np.int64)
    phone=[]
    for i in range(24):
        phone.extend([phoneno[i]]*freq[i])
    np.random.shuffle(phone)
    return phone[:100]

def suspect_imei():
    ime=np.random.randint(100000000000000,999999999999999,3,dtype=np.int64)
    imei1=np.random.choice(ime,100)
    return imei1

def other_imei(imei2):
    imei2=np.array(imei2)
    imei2=imei2*100000+imei2
    for i in range(7):
        imei2[i]=imei2[i]+99
    return imei2

def typec():
    return np.random.choice(['CALLIN','CALLOUT','SMSIN','SMSOUT'],100)

def time(typec):
    start=datetime.datetime(2005,3,25,0,0,0)
    start=start+datetime.timedelta(days=np.random.randint(0,4000))
    end=datetime.datetime.now()
    delta=end-start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = np.random.randint(0,int_delta,100)
    random_second.sort()
    end_seconds=np.random.randint(0,90*60,100)
    startdates=[]
    enddates=[]
    for i in range(100):
        st=start+datetime.timedelta(seconds=int(random_second[i]))
        startdates.append(st)
        if(typec[i] in ['SMSIN','SMSOUT']):
            enddates.append(st)
        else:
            enddates.append(st+datetime.timedelta(seconds=int(end_seconds[i])))
    return startdates,enddates    

def tower():
    return np.random.randint(10000000,99999999,100)

def location():
    city=np.array([[28,77],[19,72],[33,75],[30,69],[33,42],[77,30],[35,-95],[25,55],[-25,130],[-8,34]])
    idx=np.random.randint(0,10,5)
    loclat=[]
    lonlat=[]
    for i in range(5):
        lat=np.random.normal(city[idx[i]][0],(idx[i]+1)/2.0,20)
        lon=np.random.normal(city[idx[i]][1],(idx[i]+1)/2.0,20)
        for j in range(20):
            loclat.append(round(lat[j],6))
            lonlat.append(round(lon[j],6))
    return loclat,lonlat

def operators():
    operators=['Airtel','BSNL','Jio','Vodafone','Idea']
    return np.random.choice(operators,100)

def imsi(phone):
    front=np.random.randint(10000,99999,100)
    front=front*10000000000
    front+=phone
    return front

def main():
    df=pd.DataFrame();
    suspect=suspect_no;
    df['PHONE']=phone_no();
    df['IMEI1']=suspect_imei();
    df['IMEI2']=other_imei(df['PHONE']);
    df['TYPE']=typec();
    df['START'],df['END']=time(df['TYPE']);
    df['IMSI']=imsi(df['PHONE'])
    df['TOWER']=tower();
    df['LATITUDE'],df['LONGITUDE']=location();
    df['OPERATORS']=operators();
    df.to_csv('fourth.csv',index=False)
main()
```
